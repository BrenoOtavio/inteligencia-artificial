```
from math import sqrt
import pprint
#Abre arquivo
maze = ''
with open('maze01.txt', 'r') as fp:
    maze = fp.readlines()
    tam_linha = len(maze[1])

#Trasforma em matriz
for i in range(len(maze)):
  maze[i] = list(maze[i])

#Acha a entrada e a saída
for i in range(0, len(maze)):
  for j in range(0, tam_linha):
    if maze[i][j] == 'E':
      inicio = [i,j]

    if maze[i][j] == 'S':
      saida = [i, j]
      break
```

```
#calcula a distancia entre dois pontos
def heuristica(posicao):
  xA = posicao[0]
  yA = posicao[1]
  xB = saida[0]
  yB = saida[1]
  return (sqrt((xA-xB)**2) + ((yA-yB)**2))
```

```
#função de vericar melhor heuristiva
def melhor_heuristica(a, b, c,d):
  if a<b and a<c and a<d:
    return True
  else: return False
```

```
visitados = []
acoes = []
bifurcacoes = []
def busca(agente):
  #funções de movimentação
  mov_direita = lambda x : x[1] + 1
  mov_esquerda = lambda x : x[1] - 1
  mov_cima = lambda x : x[0] - 1
  mov_baixo = lambda x : x[0] + 1

  while True: 

    #verifica se esta proximo da saida  
    if maze[agente[0]][agente[1]+1] == 'S' or maze[agente[0]+1][agente[1]] == 'S' or maze[agente[0]][agente[1]-1] == 'S' or maze[agente[0]-1][agente[1]] == 'S':
      print('Achou a saida')
      acoes.append('SAIDA')
      return

    #verifica caminho para direita
    if maze[agente[0]][agente[1]+1] == ' ' and [agente[0], agente[1]+1] not in visitados:
      prox = [agente[0], agente[1]+1]
      #calcula a distancia ate a saida
      irDireita = heuristica(prox)
      #variavel de caminho disponivel caso haja uma bifurcação
      dir_disponivel = 1
    else: 
      #Caso o caminho não esteja disponivel adiciona uma distancia grande para a comparação
      irDireita = 999999
      dir_disponivel = 0
      
    #verifica caminho para baixo
    if maze[agente[0]+1][agente[1]] == ' ' and [agente[0]+1, agente[1]] not in visitados:
      prox = [agente[0]+1, agente[1]]
      irBaixo = heuristica(prox)
      baix_disponivel = 1
    else: 
      irBaixo = 999999
      baix_disponivel = 0

    #verifica caminho para esquerda
    if maze[agente[0]][agente[1]-1] == ' ' and [agente[0], agente[1]-1] not in visitados:
      prox = [agente[0], agente[1]-1]
      irEsquerda = heuristica(prox)
      esq_disponivel = 1
    else: 
      irEsquerda = 999999
      esq_disponivel = 0

    #verifica caminho para cima
    if maze[agente[0]-1][agente[1]] == ' ' and [agente[0]-1, agente[1]] not in visitados:
      prox = [agente[0]-1, agente[1]]
      irCima = heuristica(prox)
      cima_disponivel = 1
    else: 
      irCima = 999999
      cima_disponivel = 0
    
    #armazena as bifurcacoes para voltar caso va a um beco sem saida
    if (dir_disponivel+baix_disponivel+esq_disponivel+cima_disponivel) >= 2:
      bifurcacoes.append([agente[0], agente[1]])

    #Verifica qual caminho tem a menor distancia ate a saida e move o agente
    #move direita
    if melhor_heuristica(irDireita, irBaixo, irEsquerda, irCima):
      agente[1] = mov_direita(agente)
      acoes.append('direita')

    #move baixo
    if melhor_heuristica(irBaixo, irCima, irDireita, irEsquerda):
      agente[0] = mov_baixo(agente)
      acoes.append('baixo')

    #move cima
    if melhor_heuristica(irCima, irBaixo, irDireita, irEsquerda):
      agente[0] = mov_cima(agente)
      acoes.append('cima')
    
    #move esquerda
    if melhor_heuristica(irEsquerda, irDireita, irCima, irBaixo):
      agente[1] = mov_esquerda(agente)
      acoes.append('esquerda')

    #verifica se chegou em um beco sem saida
    if dir_disponivel+baix_disponivel+esq_disponivel+cima_disponivel == 0 :
      if(len(bifurcacoes)) == 0:
        print('Erro para achar a saida')
        return

      aux = agente
      #retorna ate a ultima bifurcacao
      agente = bifurcacoes.pop()
      
      acoes.append('De: ' + str(aux) + ' volta para: ' + str(agente))
      
    
    #armazena os caminhos visitados
    visitados.append([agente[0], agente[1]])
```

```
busca(inicio)

print('Posicoes visitadas: ',len(visitados))

aux_visi = []
aux_acoes = []

#excluindo caminhos que não levam ate a saída
for i in range(len(visitados)):
  if visitados[i] in aux_visi:
    while True:
      aux_acoes.pop()
      if aux_visi.pop() == visitados[i]:
        break
  aux_visi.append(visitados[i])
  aux_acoes.append(acoes[i])
aux_acoes.append(acoes[-1])

print('Caminho ate a saida')
pprint.pprint(aux_acoes)
```
